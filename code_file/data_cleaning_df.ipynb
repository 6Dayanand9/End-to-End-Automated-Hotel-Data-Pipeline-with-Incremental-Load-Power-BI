```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import os

pd.set_option('display.max_columns', None)
```

```python
base_path = Path().resolve().parent
file_path = base_path / "current_raw_data" / "combined_data.csv"

df = pd.read_csv(file_path)
```

# **Data Cleaning** 

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Closed On?'],inplace=True)
```

```python
df.shape
```

```python
new_columns = []

for col in df.columns:
    col = col.strip()        # remove leading & trailing spaces
    col = col.lower()        # convert to lowercase
    col = col.replace(' ', '_')  # replace spaces with underscore
    new_columns.append(col)

df.columns = new_columns
```

```python
df.columns
```

```python
df.rename(columns={
    'delivery_time_(in_mins)': 'delivery_time_in_mins',
    'discounts_offered_(%)': 'discounts_offered_pct',
    'average_rating.1': 'average_rating_1',
    'competitors_in_the_area_(count)': 'competitors_in_the_area_count'
}, inplace=True)
```

```python
df.rename(columns={'order_status?':'order_status'},inplace=True)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df[['restaurant_type','restaurant_type.1']]
```

```python
df.drop(columns=['restaurant_type.1'],inplace=True)
```

```python
df.drop(columns=['customer_name'],inplace=True)
```

```python
df.isnull().sum()
```

# **Fixing Data Types**

```python
df['order_delivered_date'].head()
```

```python
df['order_delivered_date'] = pd.to_datetime(df['order_delivered_date'], format='mixed', errors='coerce')
```

```python
df['order_delivered_date'].info()
```

```python
df.head()
```

```python
df[['time_of_order','time_of_delivery']].head()
```

```python
for col in ['time_of_order', 'time_of_delivery']:
    df[col] = (
        pd.to_timedelta(df[col], unit='D')
        .dt.round('1s')
        .astype(str)
        .str.split().str[-1]
    )
```

```python
df[['time_of_order','time_of_delivery']].head()
```

```python
df[df['time_of_order'] > df['time_of_delivery']].shape[0]
```

```python
df = df[(df['time_of_order']) <= (df['time_of_delivery'])]
```

```python
df.shape
```

```python
df = df[
    (df['time_of_delivery'] - df['time_of_order']) <= pd.Timedelta(hours=2)
]
```

```python
df.shape
```

```python
df.info()
```

```python
df.sample(5)
```

```python
for col in ['time_of_order', 'time_of_delivery']:
    df[col] = (
        df[col]
        .astype(str)
        .str.split()
        .str[-1]
    )
```
