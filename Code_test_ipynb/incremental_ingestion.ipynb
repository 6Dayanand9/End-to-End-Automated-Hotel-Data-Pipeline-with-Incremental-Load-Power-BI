```python

# -------------------------------
# 1. Read CSV
# -------------------------------
base_path = Path().resolve()
file_path = base_path / "current_clean_data" / "clean_data.csv"

df = pd.read_csv(file_path)

# -------------------------------
# 2. MySQL Connection
# -------------------------------
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="123456",
    database="food_data_model"
)

cursor = conn.cursor()

# -------------------------------
# 3. Create Dynamic Insert Query
# -------------------------------
columns = ", ".join([f"`{col}`" for col in df.columns])   # safe column names
placeholders = ", ".join(["%s"] * len(df.columns))

query = f"INSERT INTO fact_orders ({columns}) VALUES ({placeholders})"

# -------------------------------
# 4. Convert DataFrame → Tuples
# -------------------------------
data = [tuple(row) for row in df.to_numpy()]

# -------------------------------
# 5. Batch Insert (FAST 🚀)
# -------------------------------
batch_size = 1000

for i in range(0, len(data), batch_size):
    batch = data[i:i+batch_size]
    cursor.executemany(query, batch)
    conn.commit()

print("✅ Data inserted successfully 🚀")

# -------------------------------
# 6. Close Connection
# -------------------------------
cursor.close()
conn.close()
```
