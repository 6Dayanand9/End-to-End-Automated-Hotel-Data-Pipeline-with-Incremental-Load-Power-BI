# **Data Cleaning**

```python
df.drop(columns=['year','month'],inplace=True)
```

```python
new_columns = []

for col in df.columns:
    col = col.strip()        # remove leading & trailing spaces
    col = col.lower()        # convert to lowercase
    col = col.replace(' ', '_')  # replace spaces with underscore
    new_columns.append(col)

df.columns = new_columns
```

```python
df.rename(columns={
    'delivery_time_(in_mins)': 'delivery_time_in_mins',
    'discounts_offered_(%)': 'discounts_offered_pct',
    'average_rating.1': 'average_rating_1',
    'competitors_in_the_area_(count)': 'competitors_in_the_area_count'
}, inplace=True)
```

```python
df = df.drop_duplicates()
```

# **Fixing Data Types**

```python
df['order_delivered_date'].head()
```

```python
df['order_delivered_date'] = pd.to_datetime(df['order_delivered_date'], format='mixed', errors='coerce')
```

```python
df['order_delivered_date'].info()
```

```python
df[['time_of_order','time_of_delivery']].head()
```

```python
df[df['time_of_order'] > df['time_of_delivery']].shape[0]
```

```python
import pandas as pd

# ----------------------------
# 1. Convert to timedelta (correct way for string time)
# ----------------------------
for col in ['time_of_order', 'time_of_delivery']:
    df[col] = pd.to_timedelta(df[col]).dt.round('1s')

# ----------------------------
# 2. Remove invalid rows
# ----------------------------
df = df[df['time_of_order'] <= df['time_of_delivery']]

# ----------------------------
# 3. Filter <= 2 hours gap
# ----------------------------
df = df[
    (df['time_of_delivery'] - df['time_of_order']) <= pd.Timedelta(hours=2)
]

# ----------------------------
# 4. Convert to HH:MM:SS (for MySQL)
# ----------------------------
for col in ['time_of_order', 'time_of_delivery']:
    df[col] = df[col].astype(str).str.split().str[-1]

# ----------------------------
# 5. Final check
# ----------------------------
print(df[['time_of_order','time_of_delivery']].head())
print("Final shape:", df.shape)
```

```python
df.sample(5)
```

```python
# ----------------------------
# 1. Create date_id (YYYYMMDD)
# ----------------------------
df['date_id'] = df['order_delivered_date'].dt.strftime('%Y%m%d').astype(int)

# ----------------------------
# 2. Move date_id to first column
# ----------------------------
cols = ['date_id'] + [col for col in df.columns if col != 'date_id']
df = df[cols]
# ----------------------------
# 3. Drop original date column
# ----------------------------
df.drop(columns=['order_delivered_date'], inplace=True)

# ----------------------------
# 4. Check
# ----------------------------
df.head()
```

```python
# now save this data to current_clean_data folder overwrite old data

import os

file_path = "clean_data.csv"

# Save and overwrite
df.to_csv(file_path, index=False)

print(f"✅ Data saved successfully at: {file_path}")
```
